```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Paths (adjust to your folders)
face_dir = "../../Dataset/ProcessedFusionFull/Drowsy/face"
dotmap_dir = "../../Dataset/ProcessedFusionFull/Drowsy/dotmap"
landmark_dir = "../../Dataset/ProcessedFusionFull/Drowsy/landmark"

# Get 4 sample base filenames
samples = os.listdir(face_dir)[:4]

plt.figure(figsize=(12, 8))
for i, base_name in enumerate(samples):
    base = os.path.splitext(base_name)[0]

    # Load data
    face = cv2.imread(os.path.join(face_dir, base + ".png"), cv2.IMREAD_GRAYSCALE)
    dotmap = cv2.imread(os.path.join(dotmap_dir, base + ".png"), cv2.IMREAD_GRAYSCALE)
    landmarks = np.load(os.path.join(landmark_dir, base + ".npy"))

    # Plot
    plt.subplot(4, 3, i * 3 + 1)
    plt.imshow(face, cmap='gray')
    plt.title(f"Face {i+1}")
    plt.axis('off')

    plt.subplot(4, 3, i * 3 + 2)
    plt.imshow(dotmap, cmap='hot')
    plt.title("Dotmap")
    plt.axis('off')

    plt.subplot(4, 3, i * 3 + 3)
    plt.imshow(face, cmap='gray')
    for (x, y) in landmarks:
        plt.plot(x * face.shape[1], y * face.shape[0], 'bo', markersize=2)
    plt.title("Landmarks")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# ✅ Use your actual path here:
raw_dir = "../../Dataset/Raw/Drowsy"  # Adjust as needed

# Load 4 sample filenames
samples = os.listdir(raw_dir)[:3]

plt.figure(figsize=(10, 5))
for i, fname in enumerate(samples):
    img_path = os.path.join(raw_dir, fname)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(1, 4, i + 1)
    plt.imshow(img)
    plt.title(f"Raw {i+1}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
